```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data_dir="/Users/laurasansc/github/algorithms_bioinformatics/data/"
```

```python
# Read the input file
raw_data_file = data_dir + "raw/predictions_smm.txt"
raw_data = pd.read_csv(raw_data_file, sep="\t")

raw_data
```

```python
# Data distribution by IC50 and species
#raw_data['ic50'].hist(by=raw_data['species'])
```

```python
# Group by human
human_data = raw_data[raw_data['species'].isin(['human'])]

human_data
```

```python
#filtered_data = human_data.groupby("allele").filter(lambda x: (x["ic50"] < 500))
filtered_data = human_data.loc[human_data['ic50'] < 500 ]

filtered_data
```

```python
# get unique alleles names
alleles = filtered_data["allele"]
alleles = alleles.drop_duplicates()

# get top 20
alleles = alleles.head(20)
list_alleles = alleles.to_list()
```

```python
# Get a final data frame with 20 human selected HLA alleles with IC50 > 500 
selected_alleles = pd.merge(filtered_data, alleles, on='allele')

selected_alleles
```

```python
# For each of the alleles
# Split 80/20 for train / test
# Write an ouput file for each

for item in list_alleles:
    allele2split = selected_alleles.loc[selected_alleles['allele']== item]
    train, test = train_test_split(allele2split, test_size=0.2)
    
    # Re-write item name without *
    item = item.replace(' ', '_')
    item = item.replace('*', '_')
    
    # write csv files
    name_file_train = item + "_train.csv"
    train.to_csv("./data/"+name_file_train)
    
    name_file_test = item + "_test.csv"
    test.to_csv("./data/"+name_file_test)
```
